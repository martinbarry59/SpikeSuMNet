```python
def labelstostring(labels):
    out = []
    for label in labels:
        out += [''.join(label).replace('A','0').replace('B','1')]
    return out
def tostring(tensor):
    string = ''
    for i in tensor:
        string += str(i.item())
    return string
def meyniel_sim(epochs,px):
    simulation = {}
    simulation["mazes"] = [[0,1]]
    tx = torch.cat([torch.Tensor([1-px,px]).unsqueeze(0),torch.Tensor([1-px,px]).unsqueeze(0)])
    simulation["transitions"] = [tx]
    simulation["epochs"] = epochs
    simulation["number_rooms"] = 2
    simulation["Dirichlet"] = 0
    simulation["volatility"] = 0
    simulation["n_moves"] = 2
    simulation["change_points"] = {}
    simulation["change_points"][0] = 0
    simulation["rooms"] = [0]
    simulation["maze"] = [0]
    current_maze = 0
    for i in range(epochs):
        x = 0 + 1 * (px > torch.rand(1))
        simulation["rooms"] += [x]
        simulation["maze"] += [0]
    return simulation
```

```python
def calculate_error(activity, values,labels):
    points = [1,2,3,4,5]
    
    dic = {}
    avg = torch.mean(activity[500:])
    std = torch.std(activity[500:])
    avg_data = torch.mean(values[:,0]).item()
    std_data = torch.std(values[:,0]).item()
    for point in points:
        for epoch in range(500,epochs - point):
            key = tostring(simulations[0]['rooms'][epoch:epoch + point])
            act = (torch.mean(activity[epoch+point-1:epoch + point])).item()
            dic[key] = [act] if key not in dic.keys() else dic[key] + [act]
    y =[]
    saved_labels = []
    for key in dic.keys():
        if key[-1]=='0':
            y +=[statistics.mean(dic[key])]
            saved_labels += [key]
    avg = statistics.mean(y)
    std = statistics.stdev(y)
    y = [avg_data + std_data*(data-avg)/std for data in y]
    y_save = y.copy()
    y = []
    for i in range(len(labels)):
        y +=[values[i,0].item()]
    error = 0 
    for key in dic.keys():
        if key[-1]=='0':
            error += (y[labels.index(key)]-y_save[saved_labels.index(key)])**2
    return error

```
