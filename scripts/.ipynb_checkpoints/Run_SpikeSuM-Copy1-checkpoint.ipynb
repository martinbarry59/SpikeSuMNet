```python
def activity_average_sim(datas,start = 0, end = -1,window = 100):

    x = np.array(datas).reshape(np.array(datas).shape[0],-1,window),axis=2)
    #x = np.array(datas)
    average = np.mean(np.array(datas)[start:end].reshape(-1,window),axis = 1)
    std = np.std(x[:,start:end],axis = 0)
    return list(average), list(average - std)[::-1], list(average + std)
```
