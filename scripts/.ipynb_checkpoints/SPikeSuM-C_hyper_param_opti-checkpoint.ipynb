```python
min_error = 100
import skopt
import math
# warnings.filterwarnings("ignore")
def optimise_model(param_network, simulation):
    write_file = 'logs/log_optimisation.txt'
    print('---------------------------START OF OPTIMIZATION--------------------------------------------------', file=open(write_file, "a"))
    global min_error
    min_error = 100
    SPACE = [
        skopt.space.Real(0.25, 0.32, name='beta', prior='uniform'),
        skopt.space.Real(0.7, 1.2, name='r', prior='uniform'),
        skopt.space.Real(0.01, 0.1, name='lr_msp', prior='uniform'),
        skopt.space.Real(0.15, .2, name='max_commitment', prior='uniform'),
        skopt.space.Real(0.01, 0.1, name='wta_speed', prior='uniform')]

    @skopt.utils.use_named_args(SPACE)
    def objective(**params):
        
        global min_error
        all_params = {**params}
        print(all_params, file=open(write_file, "a") )
        param_network['n_memory'] = 5
        param_network['SpikeSuM_module']['eta1'] = 1e-05
        param_network['SpikeSuM_module']['eta2'] = 0.005
        param_network['SpikeSuM_module']['theta'] = 0.35
        param_network['selector_module']['beta'] = all_params['beta']
        param_network['selector_module']['r'] = all_params['r']
        param_network['selector_module']['lr_msp'] = all_params['lr_msp']
        param_network['selector_module']['max_commitment'] = all_params['max_commitment']
        param_network['selector_module']['module_inhib'] = 10
        param_network['selector_module']['wta_speed'] = all_params['wta_speed']
        param_network['SpikeSuM_module']['N'] = 50
        utils.nice_print(param_network)
        net = network.SpikeSuMM(param_network,None)
        _, epoch = network.full_simulation(simulations,net)
        error = torch.mean(net.info["error"]).item() + (epochs - epoch)
        prefix ='tania_task'

        if math.isnan(error):
            error = 100
        if error < min_error:
            print('Min error : {0} \n'.format(error), file=open(write_file, "a"))

            min_error = error
#             utils.save(net.info, file = prefix, type_= 'data')
            print(net.info['error'].shape)
            utils.plotly_plot([net.info['error']],['time step','Error(t)'],prefix+'_Error')


        return error
    try:
        
        filename = 'results/data/tania_task.pkl'
        results = skopt.load(filename)
        results = skopt.gp_minimize(objective, SPACE, n_jobs = -1, noise=1e-10, x0 = results.x,y0= results.fun)
    except:
        
        results = skopt.gp_minimize(objective, SPACE)
    
    skopt.dump(results,'results/data/tania_task.pkl',store_objective=False)

    print(model,': min error of:', results.fun,' for values:', results.x, file=open(write_file, "a"))
    print('---------------------------END OF OPTIMIZATION--------------------------------------------------', file=open(write_file, "a"))
```
